##  use networkx  and/or create map of the social network

```python
# import the python modules needed
import json
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.image as mpimg
```

```python
# plotting the networkx graph
# set the dataframe
snadf = df1

# set the node to start
node = 1
fig=plt.figure(figsize=(10,10))
ax=plt.subplot(111)
ax.set_aspect('equal')


# initialize the graph
G=nx.Graph()

# Iterate through the rows of the dataframe df1 and add edges to connect the nodes
for each in df1.iterrows():   
    
    # Add nodes and edges    
    #G.add_node(each[1]['NAME'])
    #G.add_node(each[1]['NAME'], image=img1)
    G.add_edge(each[1]['NAME'], each[1]['CONNECTED'])     
    
    # check to see if you are at the end of the rows of data frame df1
    if node < 9:
     node=node+1 
     


nx.draw(G, with_labels = True) 
```
